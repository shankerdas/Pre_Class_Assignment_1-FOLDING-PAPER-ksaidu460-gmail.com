### Problem 1

```python
"""
Code to calculate the thickness when the paper is folded 43 times
"""
THICKNESS = 0.00008
folded_paper=43
folded_thickness = THICKNESS*(2**folded_paper)
print("Thickness: {}meters".format(folded_thickness))
```

### Problem 2

```python
"""
Convert meters to kilometers
"""
THICKNESS =0.00008
folded_paper=43
folded_thickness = THICKNESS*(2**folded_paper)
print("Thickness: {:.2f}0,000 km".format(folded_thickness/1000))
```

```python
"""
View the chart. Title and axis label naming。
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list,linewidth=5,color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-np.pi/2, np.pi/2, 31)
y = np.cos(x)**3

# 1) remove points where y > 0.7
x2 = x[y <= 0.7]
y2 = y[y <= 0.7]

# 2) mask points where y > 0.7
y3 = np.ma.masked_where(y > 0.7, y)

# 3) set to NaN where y > 0.7
y4 = y.copy()
y4[y3 > 0.7] = np.nan

plt.plot(x*0.1, y, 'o-', color='blue', label='folded_paper')
plt.plot(x2*0.4, y2, 'o-', color='red', label='folded_per ms')
plt.plot(x*0.7, y3, 'o-', color='green', label='paper_thickness')
plt.plot(x*1.0, y4, 'o-', color='black', label='total_folded')
plt.legend()
plt.title('folded_thickness')
plt.show()
```
